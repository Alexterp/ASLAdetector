```python
import numpy as np
import matplotlib.pyplot as plt
import math
import plotly
```

```python
# Threshold function
x1 = np.arange(-4,0,0.01)
x2 = np.arange(0,4,0.011)

y1 = 0*x1
y2 = 1 + 0*x2

plt.grid(True)
plt.axhline(0.5,color='black') # x = 0
plt.axvline(0,color='black') # y = 0
plt.plot(x1,y1,'blue')
plt.plot(x2,y2,'blue')

plt.show()
```

```python
#SIGMOID

x = np.arange(-4,4,0.01)

y = 1/(1+ np.exp(-x))

plt.grid(True)
plt.axhline(0.5,color='black') # x = 0
plt.axvline(0,color='black') # y = 0
plt.plot(x,y,'blue')
```

```python
# ReLU

x1 = np.arange(-4,0,0.01)
x2 = np.arange(0.01,4,0.01)

y1 = 0*x1
y2 = x2

plt.grid(True)
plt.axhline(0,color='black') # x = 0
plt.axvline(0,color='black') # y = 0
plt.plot(x1,y1,'blue')
plt.plot(x2,y2,'blue')

plt.show()
```
